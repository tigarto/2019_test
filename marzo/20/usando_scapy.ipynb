```python
num_packet_in = 0
num_packet_out = 0
for p in a:
    if  p.getlayer('OFPTPacketIn') != None and p.getlayer('OFPTPacketIn').summary() == 'OFPTPacketIn':
        num_packet_in += 1
    elif p.getlayer('OFPTPacketOut') != None and p.getlayer('OFPTPacketOut').summary() == 'OFPTPacketOut':
        num_packet_out += 1
        

print("\nNumero de muestras: %s"%(len(a)))
print("Arribo de la primera muestra: %f"%(a[0].time))
print("Arribo de la primera muestra: %f"%(a[len(a)-1].time))
print("Tiempo de muestreo: %f"%(a[len(a)-1].time - a[0].time))

print("Numero de packets_in: %d" %(num_packet_in))
print("Numero de packets_out: %d"%(num_packet_out))

    
```

```python
num_pkt_per_sec = []
sec_act = 1
pps_act = 0
for p in a:
    t_p_act = p.time - a[0].time
    if t_p_act <= sec_act:
        pps_act += 1
    else:
        num_pkt_per_sec.append(pps_act)
        sec_act += 1
        pps_act = 1
        
print(num_pkt_per_sec)        
    
```

```python
num_pkt_in_per_sec = []
sec_act = 1
pps_in_act = 0
for p in a:
    t_p_act = p.time - a[0].time
    if  p.getlayer('OFPTPacketIn') != None and p.getlayer('OFPTPacketIn').summary() == 'OFPTPacketIn':
        if t_p_act <= sec_act:
            pps_in_act += 1
        else:
            num_pkt_in_per_sec.append(pps_in_act)
            sec_act += 1
            pps_in_act = 1
        
print(num_pkt_in_per_sec)        
```
